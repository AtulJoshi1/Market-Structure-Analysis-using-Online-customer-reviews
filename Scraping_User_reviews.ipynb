```
from bs4 import BeautifulSoup
import ssl
import re
import pandas as pd
```

```
# For ignoring SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```
import httplib2
from bs4 import  SoupStrainer
hyperlinks=[]

http = httplib2.Http()
status, response = http.request('https://www.dpreview.com/reviews/')
```

```
soup = BeautifulSoup(response, 'html.parser')
```

```
print(soup.prettify())
```

```
hyperlinks=[]
```

```
for i in soup.find_all('a'):
  hyperlinks.append(i.get('href'))
```

```
hyperlinks[:100]
```

```
links1=[]
for i in hyperlinks:
    matchObj = re.match( r'https://www.dpreview.com/reviews/',i, re.M|re.I)
    comp = re.match(r'https://www.dpreview.com/reviews/buying',i,re.M|re.I)
    if(comp):
        continue
    if(matchObj):
        links1.append(i)
```

```
links1
```

```
unique_links = list(set(links1))
```

```
len(unique_links)
```

```
reviews=[]
for i in unique_links:
  soup1 = BeautifulSoup(http.request(i)[1],'html.parser')
  reviews.append(soup1.p)
```
